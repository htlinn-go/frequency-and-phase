```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

# Streckenidentifikation

## Eingang - Sprung

```python
t1 = np.arange(-1,0,0.01)
t2 = np.arange(0,5,0.01)
xe1 = np.linspace(0,0,len(t1))
xe2 = np.linspace(1,1,len(t2))
t = np.concatenate((t1,t2))
xe = np.concatenate((xe1,xe2))
```

```python
Kp = 1
tau = 1
xa = Kp*(1-np.exp(-t/tau))*xe
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(t,xe)
ax1.set_xlim(-1, 5)
ax1.set_ylim(-0.25, 1.7)
ax1.set_xlabel('time')
ax1.set_ylabel('xe(t)')
ax1.grid(True)

ax2.plot(t,xa)
ax2.set_xlim(-1, 5)
ax2.set_ylim(-0.25, 1.7)
ax2.set_xlabel('time')
ax2.set_ylabel('xa(t)')
ax2.grid(True)
```
