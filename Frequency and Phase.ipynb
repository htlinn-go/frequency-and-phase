# Frequency and Phase

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath as cm
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

## Function definitions

```python
def xc(w,C):
    return 1/(1j*w*C)

def xl(w,L):
    return 1j*w*L

def prntZs(ws,Zs):
    for i , Z in enumerate(Zs):
        print('Frequency = ' + str(np.round(ws[i])) + ' rad/s')
        print('Z = ' + str(np.round(Z,2)) + ' Ohm')
        print('Magnitude = ' + str(np.round(abs(Z))) + ' Ohm')
        print('Phase = ' + str(np.round(cm.phase(Z)*180/np.pi)) + '°')
        print('\n')

def pltCs(Zs,ws,scale):
    
    ph = []
    M = []
    for i, Z in enumerate(Zs):
        ph.append(cm.phase(Z)*180/np.pi)
        M.append(abs(Z))
        
    fig1 = plt.figure()
    ax = fig1.add_subplot(2, 1, 1)
    ax.plot(ws,M)
    if scale == 'log':
        ax.set_xscale('log')
        ax.set_yscale('log')    
    ax.grid()
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency in rad/s')
    
        
    fig2 = plt.figure()
    ax = fig2.add_subplot(2, 1, 1)
    ax.plot(ws,ph)
    if scale == 'log':
        ax.set_xscale('log')   
    ax.grid()
    plt.ylabel('Phase in °')
    plt.xlabel('Frequency in rad/s')
    
    
```

## Capacitor

```python
#Frequency definition
ws=np.logspace(0,6,num=7,base=10)# 1/s
#Capacitor values
C=10**(-6) #F

Zs = xc(ws,C)  

prntZs(ws,Zs)

pltCs(Zs,ws,'log')


    
```

## Inductance

```python
#Frequency definition
ws=np.logspace(0,6,num=7,base=10)# 1/s
# values
L=10**(-3) #H

Zs = xl(ws,L)  

prntZs(ws,Zs)

pltCs(Zs,ws,'log')
```

## RLC

```python
#Frequency definition
ws=np.logspace(0,6,num=7,base=10)# 1/s
#values
L=10**(-3) #H
C=10**(-6) #F
R=10**3 #Ohm


Zs = R+ xc(ws,C) + xl(ws,L)  

prntZs(ws,Zs)

pltCs(Zs,ws,'log')
```
