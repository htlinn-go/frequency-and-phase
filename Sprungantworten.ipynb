# Sprungantworten für verschiedene Strecken

## Initialisierung

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

## Eingang - Sprung

```python
t1 = np.arange(-1,0,0.01)
t2 = np.arange(0,5,0.01)
xe1 = np.linspace(0,0,len(t1))
xe2 = np.linspace(1,1,len(t2))
t = np.concatenate((t1,t2))
xe = np.concatenate((xe1,xe2))
```

## Proportionales Verhalten 

```python
Kp = 1.5
xa = xe*Kp
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(t,xe)
ax1.set_xlim(-1, 5)
ax1.set_ylim(-0.25, 1.7)
ax1.set_xlabel('time')
ax1.set_ylabel('xe(t)')
ax1.grid(True)

ax2.plot(t,xa)
ax2.set_xlim(-1, 5)
ax2.set_ylim(-0.25, 1.7)
ax2.set_xlabel('time')
ax2.set_ylabel('xa(t)')
ax2.grid(True)
```

## Verzögerungsglied erster Ordnung PT1

```python
Kp = 1
tau = 1
xa = Kp*(1-np.exp(-t/tau))*xe
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(t,xe)
ax1.set_xlim(-1, 5)
ax1.set_ylim(-0.25, 1.7)
ax1.set_xlabel('time')
ax1.set_ylabel('xe(t)')
ax1.grid(True)

ax2.plot(t,xa)
ax2.set_xlim(-1, 5)
ax2.set_ylim(-0.25, 1.7)
ax2.set_xlabel('time')
ax2.set_ylabel('xa(t)')
ax2.grid(True)
```

## Verzögerungsglied zweiter Ordnung - stabil

```python
Kp = 1
tau1 = 1
tau2 = 1
xa = Kp*(1-np.exp(-t/tau1)-t*np.exp(-t/tau2))*xe
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(t,xe)
ax1.set_xlim(-1, 5)
ax1.set_ylim(-0.25, 1.7)
ax1.set_xlabel('time')
ax1.set_ylabel('xe(t)')
ax1.grid(True)

ax2.plot(t,xa)
ax2.set_xlim(-1, 5)
ax2.set_ylim(-0.25, 1.7)
ax2.set_xlabel('time')
ax2.set_ylabel('xa(t)')
ax2.grid(True)
```

## Verzögerungsglied zweiter Ordnung - instabil

```python
xa = (1.5-np.exp(-t*0.708)*(0.984*np.sin(1.0793*t)+1.5*np.cos(1.0793*t)))*xe
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(t,xe)
ax1.set_xlim(-1, 5)
ax1.set_ylim(-0.25, 1.7)
ax1.set_xlabel('time')
ax1.set_ylabel('xe(t)')
ax1.grid(True)

ax2.plot(t,xa)
ax2.set_xlim(-1, 5)
ax2.set_ylim(-0.25, 1.7)
ax2.set_xlabel('time')
ax2.set_ylabel('xa(t)')
ax2.grid(True)
```

## integrierendes Verhalten

```python
Ki = 0.2
xa = Ki*np.cumsum(xe)*(t[1]-t[0])
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(t,xe)
ax1.set_xlim(-1, 5)
ax1.set_ylim(-0.25, 1.7)
ax1.set_xlabel('time')
ax1.set_ylabel('xe(t)')
ax1.grid(True)

ax2.plot(t,xa)
ax2.set_xlim(-1, 5)
ax2.set_ylim(-0.25, 1.7)
ax2.set_xlabel('time')
ax2.set_ylabel('xa(t)')
ax2.grid(True)
```

## Differenzierendes verhalten

```python
Kd = 1
xa = Kd*np.diff(xe,prepend=0)/(t[1]-t[0])
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(t,xe)
ax1.set_xlim(-1, 5)
ax1.set_ylim(-0.25, 1.7)
ax1.set_xlabel('time')
ax1.set_ylabel('xe(t)')
ax1.grid(True)

ax2.plot(t,xa)
ax2.set_xlim(-1, 5)
#ax2.set_ylim(-0.25, 1.7)
ax2.set_xlabel('time')
ax2.set_ylabel('xa(t)')
ax2.grid(True)
```

## Totzeitverhalten

```python
Tt = 1
xa = np.pad(xe,(int(Tt/(t[1]-t[0])),0),'edge')[0:600]
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(t,xe)
ax1.set_xlim(-1, 5)
ax1.set_ylim(-0.25, 1.7)
ax1.set_xlabel('time')
ax1.set_ylabel('xe(t)')
ax1.grid(True)

ax2.plot(t,xa)
ax2.set_xlim(-1, 5)
#ax2.set_ylim(-0.25, 1.7)
ax2.set_xlabel('time')
ax2.set_ylabel('xa(t)')
ax2.grid(True)
```

## PID Verhalten

```python
Kp = 1
Ki = 0.1
Kd = 0.1
xa = Kp*xe + Ki*np.cumsum(xe)*(t[1]-t[0]) + Kd*np.diff(xe,prepend=0)/(t[1]-t[0])
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(t,xe)
ax1.set_xlim(-1, 5)
ax1.set_ylim(-0.25, 1.7)
ax1.set_xlabel('time')
ax1.set_ylabel('xe(t)')
ax1.grid(True)

ax2.plot(t,xa)
ax2.set_xlim(-1, 5)
#ax2.set_ylim(-0.25, 1.7)
ax2.set_xlabel('time')
ax2.set_ylabel('xa(t)')
ax2.grid(True)
```
