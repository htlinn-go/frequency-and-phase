Multimeter:


```python
U_q=10 #V
R =10**6 #Ohm
R_iV=10**7 #Ohm
```

# Ermittlung des Stromes ohne Messgeräte

```python
I_R = U_q/R
print(str(I_R*10**3) + ' mA')
```

Damit kann aus der Tabelle der richtige messbereich gewählt werden. Daraus folgt der Innenwiderstand des Amperemeters zu:

```python
R_iA=10**4 # Ohm
```

# Ermittlung der Spannung ohne Messgeräte


```python
U_R=U_q
```

# Spannungsrichtiges Messen


## Ermittlung des Stromes und der Spannung

```python
R_p=1/(1/R+1/R_iV)
R_ges = R_iA+R_p
print(str(R_ges) + ' Ohm')
```

```python
I=U_q/R_ges
print(str(I*10**3) + ' mA')
```

```python
U_R=I*R_p
print(str(U_R) + ' V')
```

```python
U_V = U_R
```

```python
I_R=U_R/R
print(str(I_R*10**3) + ' mA')
```

```python
I_A=I
```

## Abweichungsberechnung für die Spannungsrichtige Messung

```python
f_U=(U_R-U_V)/U_R
print(str(f_U*100) + ' %')
```

```python
f_I=(I_A-I_R)/I_R
print(str(f_I*100) + ' %')
```
